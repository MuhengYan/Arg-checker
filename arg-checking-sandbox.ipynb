```python
import pandas as pd
import html
from IPython.core.display import display, HTML
from utils import html_escape, show, colors


df = pd.read_csv('data/sampled_args.csv')
```

```python
# change the value of this variable to the row you would like to visualize, and run the following codes
row_to_visualize = 1
```

```python
idx = row_to_visualize - 1
tokens = df['text']
tags = df['tags']
```

```python
#show the row of data
df[idx:idx+1]
```

```python
#show the colored texts
example = []
for txt in colors:
    span = '<span style="background-color:rgba('
    span += str(colors[txt])
    span += ',' 
    span += str(1) 
    span += ');">' 
    span += html_escape(txt) 
    span += '</span>'
    example.append(span)
print('Legends:')
display(HTML('\n'.join(example)))
print('------------------------------------------------------------')
print('Article Content:')
show(tokens, tags, idx, colors=colors)
```
